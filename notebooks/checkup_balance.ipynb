```python
import os
from dotenv import load_dotenv
from src.execution import BinanceExecutor

def forzar_compra_local():
    print("="*50)
    print("🚀 Iniciando prueba LOCAL de BinanceExecutor")
    print("="*50)

    # 1. Cargamos el .env de tu Mac
    load_dotenv()

    # 2. Conectamos directamente las manos del bot
    executor = BinanceExecutor(
        api_key=os.getenv('BINANCE_TESTNET_API_KEY'),
        api_secret=os.getenv('BINANCE_TESTNET_SECRET'),
        bucket_name=os.getenv('AWS_S3_BUCKET_NAME'),
        testnet=True
    )

    # 3. Disparamos la función forzando los parámetros
    print("\nMODO PRUEBA: Forzando ejecución de compra (35%)...")
    executor.ejecutar_compra_con_trailing_stop(
        symbol='BTC/USDT',
        tamaño_posicion=0.35,
        stop_loss_inicial=0.03,
        trailing_activation=0.03,
        trailing_dist=0.015
    )

if __name__ == "__main__":
    forzar_compra_local()
```
